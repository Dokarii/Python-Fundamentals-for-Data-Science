```python
#funcion normal de toda la vida
def my_function(kids):
  print("The name of the child is " + kids)

my_function(("Emil"))
```

```python
#funciones que solo aceptan argumentos posicionales
def mi_funcion(kid, /):
    print("El nombre del niño es " + kid)

mi_funcion("Emil")
```

```python
#funciones que solo aceptan argumentos de palabras clave
def mi_funcion(*, kid):
    print("El nombre del niño es " + kid)
mi_funcion(kid="Emil")
```

```python
#Funciones con *args que aceptan un número arbitrario de argumentos posicionales
def my_function(*kids):
  print("The youngest child is " + kids[2])
my_function("Emil", "Tobias", "Linus")
```

```python
def my_function(*args):
  print("Type:", type(args))
  print("First argument:", args[0])
  print("Second argument:", args[1])
  print("All arguments:", args)

my_function("Emil", "Tobias", "Linus")
```

```python
#Argumentos regulares y *args juntos
#*args debe ir al final de la lista de parámetros
def saludo(saludo, *nombres):
    for nombre in nombres:
        print(saludo + " " + nombre)
saludo("Hola", "Ricardo", "Ana", "Luis")
```

```python
#Ejemplo practico de *args
def suma(*numeros):
    resultado = 0
    for numero in numeros:
        resultado += numero
    return resultado
print(suma(1, 2, 3))  # Salida: 6
print(suma(4, 5))     # Salida: 9
```

```python
#Finding the maximum value in a list using *args
def my_function(*numbers):
  if len(numbers) == 0:
    return None
  max_num = numbers[0]
  for num in numbers:
    if num > max_num:
      max_num = num
  return f'{max_num} is the maximum number.'

print(my_function(3, 7, 2, 9, 1))
```

```python
#Funciones con **kwargs que aceptan un número arbitrario de argumentos de palabras clave
def my_function(**kid):
  print("His last name is " + kid["lname"])

my_function(fname="Tobias", lname="Refsnes")
```

```python
def my_function(**myvar):
  print("Type:", type(myvar))
  print("Name:", myvar["name"])
  print("Age:", myvar["age"])
  print("All data:", myvar)

my_function(name = "Tobias", age = 30, city = "Bergen")
```

```python
#Ejemplo practico de **kwargs
def my_function(username, **details):
  print("Username:", username)
  print("Additional details:")
  for key, value in details.items():
    print(" ", key + ":", value)

my_function("emil123", age = 25, city = "Oslo", hobby = "coding")
```

```python
#Combinar *args y **kwargs
def my_function(*args, **kwargs):
    print("Positional arguments:", args)
    print("Keyword arguments:", kwargs)
my_function(1, 2, 3, name="Alice", age=30)
```

```python
def funcion(*args):
    return sum(args)

lista = [1, 2, 3, 4]
print(funcion(*lista))
```

```python
def funcion(a,b,c,d):
    return a + b + c + d

lista = [1, 2, 3, 4]
print(funcion(*lista))
```

```python
def my_function(fname, lname):
  print("Hello", fname, lname)

person = {"fname": "Emil", "lname": "Refsnes"}
my_function(**person)
```
