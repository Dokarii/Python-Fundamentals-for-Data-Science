```python
def generar_tablas(numero, limite):
    for i in range(1, limite + 1):
        print(f"{numero} x {i} = {numero * i}")
```

```python
def inicio():
    while True:
        numero = int(input("Ingrese el número para generar la tabla de multiplicar: "))
        limite = int(input("Ingrese el límite hasta donde desea multiplicar: "))
        generar_tablas(numero, limite)

        continuar = input("quiere generar otra tabla? (s/n): ")
        if continuar.lower() != "s":
            break
```

```python
inicio()
```
