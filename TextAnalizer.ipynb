```python
phrase = input("Enter a text with at least 10 words: ")
```

```python
total_characters = len(phrase)
```

```python
total_characters_without_spaces = len(phrase.replace(" ", ""))
```

```python
total_vowels = sum(phrase.lower().count(vowel) for vowel in "aeiou")
```

```python
number_of_words = phrase.count(" ") + 1
```

```python
total_spaces = phrase.count(" ")
```

```python
remove_first_word = phrase.split()[1:]
phrase_without_first_word = " ".join(remove_first_word)
```

```python
replace_spaces_with_hyphens = phrase.replace(" ", "-")
```

```python
lowercase_to_uppercase_and_vice_versa = phrase.swapcase()
```

```python
print("Text analysis results:")
print(f"Total characters: {total_characters}")
print(f"Total characters without spaces: {total_characters_without_spaces}")
print(f"Total vowels: {total_vowels}")
print(f"Number of words: {number_of_words}")
print(f"Total spaces: {total_spaces}")
print(f"Phrase without the first word: {phrase_without_first_word}")
print(f"Phrase with spaces replaced by hyphens: {replace_spaces_with_hyphens}")
print(f"Phrase with uppercase converted to lowercase and vice versa: {lowercase_to_uppercase_and_vice_versa}")
```
