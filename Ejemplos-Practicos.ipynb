```python
mi_lista = [1,2,3,4,5,6]

def cuadrado(n):
    return n * n
    
def calcular_lista(numeros):
    for numero in numeros:
        resultado = cuadrado(numero)
        print(resultado)
```

```python
calcular_lista(mi_lista)
```

```python
def solicitar_numero():
    numero = int(input("Ingrese un número entero positivo: "))
    return numero

def calcular_factorial(n):
    factorial = 1
    for i in range(1, n + 1):
        factorial *= i
    return factorial

def resultado():
    num = solicitar_numero()
    factorial = calcular_factorial(num)
    print(f"El factorial de {num} es {factorial}")

resultado()
```

```python
import pandas as pd
ventas = [120, 150, 90, 200, 210, 130, 160]
dias_semana = ['Lunes', 'Martes', 'Miercoles', 'Jueves', 'Viernes', 'Sabando', 'Domingo']
serie_semana = pd.Series(dias_semana)
serie_ventas = pd.Series(ventas)

suma_total_ventas = serie_ventas.sum()
dia_mayores_ventas = serie_ventas.max()
promedio_ventas = serie_ventas.mean()

print(f"Suma de ventas: {suma_total_ventas}")
print(f"Mayor venta: {dia_mayores_ventas}")
print(f"Promedio de ventas: {promedio_ventas}")
```
