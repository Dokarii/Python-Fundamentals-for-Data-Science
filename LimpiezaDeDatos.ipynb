```python
import pandas as pd
```

```python
data = {'idProducto': [1, 2, 3, 3], 'Cantidad': [10, None, 30, 30], 'Precio': [100, 200, None, 300]}

df = pd.DataFrame(data)
print(df)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df_eliminado = df.dropna()
print(df_eliminado)
```

```python
valores_reemplazados = df.fillna({'Cantidad': 0, 'Precio': df['Precio'].mean()})
print(valores_reemplazados)
```

```python
valores_reemplazados['Cantidad'] = valores_reemplazados['Cantidad'].astype(int)
print(valores_reemplazados)
```

```python
valores_reemplazados = valores_reemplazados.drop_duplicates() #compara filas y elimina las filas duplicadas, dejando solo una de ellas
print(valores_reemplazados)
```

```python
valores_reemplazados = df.drop_duplicates(subset='idProducto')
print(valores_reemplazados)
```

```python
data = {
    'ID': [1, 2, 3, 4, 5],
    'Producto': ['Producto A', 'Producto B', None, 'Producto D', 'Producto E'],
    'Cantidad': [10, 20, 30, None, 50],
    'Precio': [100, 200, 300, 400, None]
}

df = pd.DataFrame(data)
df.isnull().sum()
```
